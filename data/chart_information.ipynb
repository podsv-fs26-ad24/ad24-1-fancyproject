**Chart 1: Popularity distribution**
- 9,448 songs (10.5%) have a popularity score of exactly 0. This lets us believe that either they have never been streamed, or any streams they received were too few or too long ago to register in Spotify's scoring algorithm.
- Only about 3,100 songs out of 89,000 score above 70. So genuinely popular songs are quite rare in this dataset.
- The median popularity score is 33 out of 100, which means half of all songs in the dataset score 33 or below. So the typical song reaches very few listeners compared to what we'd consider a hit.

**Chart 2: Correlations with popularity**
-  None of the audio features actually correlate that strong with popularity. The highest correlation is only around −0.13. 
- The clearest pattern is that tracks with higher instrumentalness tend to score lower on popularity.
- Louder and more danceable songs are very slightly more popular, but the effect is really small.
- Features like tempo, energy, and whether a song sounds happy or sad barely correlate with popularity at all.

**Chart 3: Genre radar**
- Classical and metal look very different from each other. Classical is quiet and acoustic, metal is loud and intense. 
- Hip-hop stands out for danceability and has almost no instrumentalness.
- Pop and rock are surprisingly similar across most features. They sit right next to each other on almost every axis. 
- Jazz is calmer and more acoustic than the others but not quite as extreme as classical. 

**Chart 4: Hits vs. non-hits**
- Hits tend to be a bit louder, slightly more danceable and more energetic than non-hits. The differences however are smaller than expected.
- The most notable gap is in acousticness: hits average 0.28 vs 0.39 for non-hits — less acoustic tracks score slightly higher on average.
- What is also interesting, is that how happy a song sounds is almost identical between hits and non-hits.

**Chart 5: Mood map**
- Most genres land in the energetic half of the map regardless of whether they sound happy or sad. Only about 20 out of 113 genres are on the calm side. 
- Hip-hop is the only one of our six focus genres that sits in the happy-energetic quadrant.
- Pop and rock end up almost on top of each other. They feel different to listen to but the data doesn't really separate them. 
- The extremes are interesting: salsa and forro are the happiest genres, while death-metal and grindcore are by far the most intense.
